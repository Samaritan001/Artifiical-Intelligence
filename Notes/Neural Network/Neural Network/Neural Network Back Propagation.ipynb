## Within One Layer
### From $z^{[l]}$ to $W^{[l]}$
$\frac{\partial z_1^{[1]}}{\partial W_{11}^{[1]}} = a^{[0]}_1$ $\ \frac{\partial z_1^{[1]}}{\partial W_{12}^{[1]}} = a^{[0]}_2$
$\ \frac{\partial z_1^{[1]}}{\partial W_{1n^{\ [0]}}^{[1]}} = a^{[0]}_{n^{[0]}}$\
$~$\
$\frac{\partial z_2^{[1]}}{\partial W_{21}^{[1]}} = a^{[0]}_{1}$
$\ \frac{\partial z_2^{[1]}}{\partial W_{22}^{[1]}} = a^{[0]}_{2}$
$\ \frac{\partial z_2^{[1]}}{\partial W_{2n^{\ [0]}}^{[1]}} = a^{[0]}_{n^{[0]}}$\
$~$\
$\frac{\partial z_{n^{\ [1]}}^{[1]}}{\partial W_{n^{\ [1]}\ 1}^{[1]}} = a^{[0]}_{1}$
$\ \frac{\partial z_{n^{\ [1]}}^{[1]}}{\partial W_{n^{\ [1]}\ 2}^{[1]}} = a^{[0]}_{2}$

$~$\

## Between Layers
### From $z^{[l+1]}$ to $a^{[l]}$
#### Background Math
$\frac{\partial}{\partial x_1}f[g_1(x_1, x_2), g_2(x_1, x_2)] = \frac{\partial f}{\partial g_1}\frac{\partial g_1}{\partial x_1} + \frac{\partial f}{\partial g_2}\frac{\partial g_2}{\partial x_1}$

#### Background Equation
$
\begin{equation}
\left\{
     \begin{array}{lr}
     z_1^{[2]} = W_{11}^{[2]}a_1^{[1]} + W_{12}^{[2]}a_2^{[1]} + \dots + W_{1n^{[1]}}^{[2]}\ a_{n^{[1]}}^{[1]} \\
     z_2^{[2]} = W_{21}^{[2]}a_1^{[1]} + W_{22}^{[2]}a_2^{[1]} + \dots + W_{2n^{[1]}}^{[2]}\ a_{n^{[1]}}^{[1]} \\
     \ \vdots \\
     z_{n^{[2]}}^{[2]} = W_{n^{[2]}\ 1}^{[2]}a_1^{[1]} + W_{n^{[2]}\ 2}^{[2]}a_2^{[1]} + \dots + W_{n^{[2]}\ n^{[1]}}^{[2]}\ a_{n^{[1]}}^{[1]}
     \end{array}
\right.
\end{equation}
$

#### Deduction
$\frac{\partial L}{\partial a_1^{[1]}} = \frac{\partial L}{\partial z_1^{[2]}}\frac{\partial z_1^{[2]}}{\partial a_1^{[1]}} + $
$\frac{\partial L}{\partial z_2^{[2]}}\frac{\partial z_2^{[2]}}{\partial a_1^{[1]}}\ +\ \dots\ +\ $
$\frac{\partial L}{\partial z_{n^{\ [2]}}^{[2]}}\frac{\partial z_{n^{\ [2]}}^{[2]}}{\partial a_1^{[1]}} = $
$\frac{\partial L}{\partial z_1^{[2]}}W_{11}^{[2]} + \frac{\partial L}{\partial z_2^{[2]}}W_{21}^{[2]} + \dots + $
$\frac{\partial L}{\partial z_{n^{\ [2]}}^{[2]}}W_{n^{[2]}\ 1}^{[2]}$

$\frac{\partial L}{\partial a_1^{[l]}} = \frac{\partial L}{\partial z_1^{[l+1]}}\ \frac{\partial z_1^{[l+1]}}{\partial a_1^{[l]}} + $
$\frac{\partial L}{\partial z_2^{[l+1]}}\ \frac{\partial z_2^{[l+1]}}{\partial a_1^{[l]}}\ +\ \dots\ +\ $
$\frac{\partial L}{\partial z_{n^{\ [l+1]}}^{[l+1]}}\ \frac{\partial z_{n^{\ [l+1]}}^{[l+1]}}{\partial a_1^{[l]}} = $
$\frac{\partial L}{\partial z_1^{[l+1]}}W_{11}^{[l+1]} + \frac{\partial L}{\partial z_2^{[l+1]}}W_{21}^{[l+1]} + \dots + $
$\frac{\partial L}{\partial z_{n^{\ [l+1]}}^{[l+1]}}W_{n^{[l+1]}\ \ 1}^{[l+1]}$

__Thus__\
$\frac{\partial L}{\partial a^{[1]}} = $
$
\begin{bmatrix}

\end{bmatrix}
$
$
\begin{bmatrix}

    \frac{\partial L}{\partial z_2^{[2]}} \\
    \vdots \\
    \frac{\partial L}{\partial z_{n^{\ [2]}}^{[2]}}
\end{bmatrix}
$ $\ =\ $
$W^{[2]^T}dz^{[2]}$
